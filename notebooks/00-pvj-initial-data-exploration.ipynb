The data contains a column `datacompleteness`, with unique values `complete` and `partial`. Let's take a look at what makes data partial.

```python
# Filter data for rows with 'datacompleteness' == 'partial'
partial_data = data[data['datacompleteness'] == 'partial']

# Calculate the sum of null values for each column and filter out nonzero values
nonzero_null_counts = partial_data.isnull().sum().loc[lambda x: x != 0].sort_values(ascending=False)

# Total number of data points (rows) in the filtered data
total_data_points = len(partial_data)

# Calculate the percentage of missing values
percent_missing = (nonzero_null_counts / total_data_points) * 100

# Create a colormap based on the percentage of missing values
cmap = plt.colormaps['viridis']
norm = plt.Normalize(vmin=percent_missing.min(), vmax=percent_missing.max())
colors = cmap(norm(percent_missing))

# Plot the bar plot
ax = nonzero_null_counts.plot(kind='barh', figsize=(15, 15), color=colors, width=0.7)
ax.axvline(total_data_points, color='r', linestyle='--', label='Total Data Points')

# Set labels and title
ax.set_xlabel('Number of Null Values')
ax.set_ylabel('Columns')
ax.set_title('Number of Null Values and Percentage of Missing Data Points')
ax.legend(loc='lower right')

# Add color bar to show the percentage scale
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, ax=ax)
cbar.set_label('Percentage of Missing Values (%)')

# Show the plot
plt.tight_layout()
plt.show()
```

We see that most features have all (or close to all) missing values. Although the partial data contains all games from the LPL, as well as from the major tournament MSI, this data will be removed from the dataset
